```python
import matplotlib.pyplot as plt
import numpy as np
# some example data to display
x = np.linspace(0,2*np.pi,400)
y = np.sin(x**2)
```

```python
# two subplot
fig, (ax1,ax2, ax3) = plt.subplots(2)
ax1.plot(x,y)
ax2.plot(x,-y, 'tab:orange')
ax2.set_title("A single plot")
ax1.set_title("A single nplot")
```

```python
# two subplot
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x,y)
axs[0, 0].set_title("A single plot")

axs[0, 1].plot(x,-y, 'tab:red')
axs[0, 1].set_title("A plot")

axs[1, 0].plot(x,y**2, 'tab:orange')
axs[1, 0].set_title("A plot M")

for ax in axs.flat:
    ax.set(xlabel='time', ylabel='y-label')
for ax in axs.flat:
    ax.label_outer()
```

```python
# two subplot

fig, axs = plt.subplots(3) # gca stands for 'get current axis'
#fig, axs = plt.subplots(3)
axs[0].plot(x,y, marker='o')
axs[0].set_title("A single plot")

axs[1].plot(x,-y, 'tab:red')
axs[2].plot(x,2*y, 'tab:orange')




for ax in axs.flat:
    ax.set(xlabel='time', ylabel='y-label')
for ax in axs.flat:
    ax.label_outer()
```
